```python
def impute_age(cols):
    Age =cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass ==2:
            return 29
        else:
            return 24
    else:
        return Age

        
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
  # dropping cabin caolumn
train.drop('Cabin',axis=1,inplace=True)
```

```python
# deal with categorical features, by creating dummy variable
```

```python
pd.get_dummies(train['Sex']) # this will cause multicollinearity issue, becuase if one is 0 other wil be definitely 1
# ML just need to know one col..having two cols will confuse the ML algo
```

```python
pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
embark = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
embark.head()
```

```python
train = pd.concat([train,sex,embark],axis=1)  #adding new cols to dataframe
```

```python
train.head() # scroll to right to see new cols are added
```

```python
# cols like ticket, name are not reqd
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
#passenger id is just the index, no use of that data
```

```python
train.drop(['PassengerId'],axis=1,inplace=True)
```

```python
#Pcalss is also a categorical columns
# we can create dummy variable for Pclass as well
```

```python
X=train.drop('Survived',axis=1)
y=train['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report # tells precision, accuracy, recall, f score etc.
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix  # confusion metricss
```

```python
confusion_matrix(y_test,predictions)
```
